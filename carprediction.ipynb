TASK2:CAR PREDICTION MODEL WITH MACHINE LEARNING

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import warnings
```

```python
url="CarPrice_Assignment.csv"
data=pd.read_csv(url)
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data=data.drop_duplicates()
```

VISUALIZING DATA

```python
plt.style.use('dark_background')
sns.set_palette('dark')
sns.histplot(data['price'])
plt.title('Distribution of Car Prices',color='white')
plt.xlabel('Price',color='White')
plt.ylabel('Count',color='White')
plt.xticks(color='White')
plt.yticks(color='White')
plt.show()
```

```python
numeric_features=['wheelbase','carlength','carwidth','curbweight',
                  'enginesize','stroke','compressionratio','horsepower',
                  'peakrpm','citympg','highwaympg','price']
correlation_matrix=data[numeric_features].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correaltion Heatmap',color='White')
plt.xticks(color='White')
plt.yticks(color='White')
plt.show()
```

PREPROCESSING DATA

```python
feature_cols=['symboling','fueltype','aspiration','doornumber','carbody',
              'drivewheel','enginelocation','wheelbase']
target_col='price'
x=data[feature_cols]
y=data[target_col]
```

```python
label_encoder=LabelEncoder()
for col in x.columns:
    if x[col].dtype=='object':
        x[col]=label_encoder.fit_transform(x[col])
        
```

SPLITTING DATA

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

FITTING THE MODEL

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
predictions = model.predict(x_test)
mse = mean_squared_error(y_test,predictions)
rmse=np.sqrt(mse)
print('Root',rmse)
```
